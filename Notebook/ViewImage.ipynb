```python
import os
import SimpleITK as sitk
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from matplotlib.pyplot import imshow
```

```python
path = '/Users/yuyu/Documents/Capstone/Data'
```

```python
inputdirs = os.listdir(path)[1:]
```

```python
inputdirs = [os.path.join(path,inputdir) for inputdir in inputdirs]
```

```python
inputdirs
```

```python
item = 'slide007_core002_classimg_nonconvex.png'
```

```python
#uint8	Unsigned integer (0 to 255)
imgs = []
for p in inputdirs:
    if os.path.isfile(os.path.join(p, item)):
            imgs.append(sitk.ReadImage(os.path.join(p, item)))
```

```python
result = sitk.MultiLabelSTAPLE(imgs, 255)
#This filter performs a pixelwise combination of an arbitrary number of input images, 
#where each of them represents a segmentation of the same scene (i.e., image).
p1_data = sitk.GetArrayFromImage(imgs[0])
```

```python
result_data = sitk.GetArrayFromImage(result)
```
